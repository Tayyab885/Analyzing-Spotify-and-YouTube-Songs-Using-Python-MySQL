#### Remove Unnecessary Columns:

```python
df.drop(['Url_spotify', 'Uri', 'Url_youtube', 'Description'],axis=1,inplace=True)
```

```python
df.head()
```

#### Handle Null Values:

```python
### Fill object columns with NA and numeric columns with 0
object_columns = df.select_dtypes(include=['object']).columns
for col in object_columns:
    df[col].fillna('NA', inplace=True)
for col in df.columns:
    if col not in object_columns:
        df[col].fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

#### Remove Duplicates:

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

#### Convert millisecond duration to minute in Duration_ms feature:

```python
df['Duration_ms']
```

```python
df['Duration_ms'] = df['Duration_ms']/60000
df['Duration_ms']
```

#### Rename the modified column to Duration_min:

```python
df = df.rename(columns = {'Duration_ms':'Duration_min'})
df.head()
```

#### Remove Irrelevent Names from Track column that starts with '?':

```python
df[df['Track'].str.startswith('?')]
```

```python
df = df[~df['Track'].str.startswith('?')]
df.head()
```

#### Calculate the Energy to Liveness ratio for each track and store it in columns 'EnergyLiveness':

```python
df['EnergyLiveness'] = df['Energy'] / df['Liveness']
df.head()
```

#### Analyze the 'views' and 'stream' columns to determine the dominant platform (YouTube or Spotify) on which a song track was most played. Create a new column called 'most_playedon' with values 'Spotify' or 'YouTube' indicating the platform with the highest play count for each song track:

```python
df['most_played']  = df.apply(lambda x: 'Spotify' if x['Stream'] > x['Views'] else 'Youtube', axis=1)
df.head()
```
